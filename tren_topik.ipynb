```python
from bertopic import BERTopic
from transformers import AutoModel, AutoTokenizer
from sentence_transformers import SentenceTransformer
from sklearn.feature_extraction.text import CountVectorizer

# Load IndoBERT-based embedding model
embedding_model = SentenceTransformer("indobenchmark/indobert-base-p1")

vectorizer_model = CountVectorizer(ngram_range=(1, 3))
                                   
# Create BERTopic model
topic_model = BERTopic(embedding_model=embedding_model, vectorizer_model=vectorizer_model, language="indonesian")

# Fit the model
topics, probs = topic_model.fit_transform(df['text_processed'])

# Visualisasi topik
topic_model.visualize_topics()
```

```python
from bertopic import BERTopic
from gensim.corpora.dictionary import Dictionary
from gensim.models.coherencemodel import CoherenceModel

# Asumsi `topic_model` sudah dilatih
# Asumsi `df['text_processed']` adalah data dokumen Anda

# 1. Ekstrak kata-kata kunci per topik
topics_words = topic_model.get_topics() 

# 2. Siapkan kamus dan korpus terlebih dahulu
tokenized_docs = [doc.split() for doc in df['text_processed']]
dictionary = Dictionary(tokenized_docs)
corpus = [dictionary.doc2bow(doc) for doc in tokenized_docs]

# 3. Filter topik-topik outlier (-1) dan filter kata kunci yang tidak ada di dalam kamus
topics_for_coherence = []
for topic_id, words in topics_words.items():
    if topic_id != -1:  # Mengecualikan topik outlier
        # Filter kata kunci yang tidak ada di kamus gensim
        filtered_words = [word for word, score in words if word in dictionary.token2id]
        if filtered_words:  # Memastikan topik tidak menjadi kosong setelah difilter
            topics_for_coherence.append(filtered_words)

# 4. Hitung coherence
coherence_model = CoherenceModel(
    topics=topics_for_coherence,
    texts=tokenized_docs,
    corpus=corpus,
    dictionary=dictionary,
    coherence='c_v'
)

coherence_score = coherence_model.get_coherence()
print(f"Skor Coherence (C_V): {coherence_score}")
```

```python
# Ekstrak kata-kata kunci per topik
# get_topics() mengembalikan dictionary dengan nomor topik sebagai key
topics_words = topic_model.get_topics()

# Buat list kosong untuk menampung semua kata kunci dari seluruh topik
all_topic_words = []

# Iterasi melalui setiap topik dan ambil kata-kata kuncinya
for topic_id, words_with_score in topics_words.items():
    # Pastikan bukan topik outlier (-1) dan topik memiliki kata kunci
    if topic_id != -1 and words_with_score:
        # Ambil hanya kata-katanya saja (bukan skornya)
        # Ambil 10 kata teratas untuk setiap topik
        top_words = [word for word, score in words_with_score[:10]]
        all_topic_words.extend(top_words)

# Hitung jumlah kata unik
num_unique_words = len(set(all_topic_words))

# Hitung jumlah total kata yang diambil (10 kata per topik non-outlier)
num_topics_without_outlier = len(topics_words) - 1
total_words = num_topics_without_outlier * 10

# Hitung skor topic diversity
topic_diversity = num_unique_words / total_words

print(f"Jumlah kata kunci unik: {num_unique_words}")
print(f"Jumlah total kata kunci: {total_words}")
print(f"Skor Topic Diversity: {topic_diversity}")
```

```python
import pickle

with open("sentiment_app/tokenizer.pkl", "rb") as f:
        tokenizer = pickle.load(f)
        print("Model dan tokenizer berhasil dimuat.")
```

```python
from keras.models import load_model

model_gru = load_model("models_fix/best_model_GRU.h5")
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

MAX_WORDS = 8786
MAX_LEN = 36

sequences_test = tokenizer.texts_to_sequences(df['text_processed'])

padded_test = pad_sequences(sequences_test, maxlen=MAX_LEN, padding='post', truncating='post')
```

```python
import numpy as np

# Prediksi menggunakan model GRU
y_pred_probs_gru = model_gru.predict(padded_test)
df['predicted_label'] = np.argmax(y_pred_probs_gru, axis=1) 

# Konversi label numerik kembali ke nama sentimen (untuk kemudahan analisis)
label_map = {0: 'netral', 1: 'positif', 2: 'negatif'}
df['predicted_label'] = df['predicted_label'].map(label_map)

# Anda juga bisa menambahkan kolom untuk menandai apakah prediksi itu benar atau salah
df['is_correct_label'] = (df['label'] == df['predicted_label'])

df['is_correct_label'].value_counts()
```

```python
import pandas as pd

df['label_num'] = df['label'].replace({'netral' : 0, 'positif' : 1, 'negatif' : 2})
df['label_num'].value_counts()
```

```python
from keras.models import load_model
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

y_pred_probs = model_gru.predict(padded_test)
y_pred = np.argmax(y_pred_probs, axis=1)

cm = confusion_matrix(df['label_num'], y_pred)
class_names = ['Netral', 'Positif', 'Negatif']

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix dari GRU')
plt.show()

print(classification_report(df['label_num'], y_pred, target_names=class_names))
```
